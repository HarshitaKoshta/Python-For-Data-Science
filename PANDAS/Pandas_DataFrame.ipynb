### Concatenate Data Frames

```python
import pandas as pd
indian_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
indian_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
#concate two dataframes
df = pd.concat([indian_weather, us_weather])
df#concate two dataframes
df = pd.concat([indian_weather, us_weather])
df
```

```python
#if you want continuous index
df = pd.concat([indian_weather, us_weather], ignore_index=True)
df
```

```python
df = pd.concat([indian_weather, us_weather],axis=1)
df
```

## Merge DataFrames

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore","hyderabad"],
    "temperature": [32,43,36,40]})
temperature_df
```

```python
humidity_df = pd.DataFrame({
    "city": ["delhi","mumbai","banglore"],
    "humidity": [68, 65, 75]})
humidity_df
```

```python
#merge two dataframes with out explicity mention index
df = pd.merge(temperature_df, humidity_df, on ='city')
df
```

```python
df = pd.merge(temperature_df, humidity_df, on ='city', how='outer')
df
```

## Numerical Indexing (.loc vs iloc)



```python
df = pd.DataFrame([1,2,3,4,5,6,7,8,9,19], index=[49,48,47,46,45, 1, 2, 3, 4, 5])
df
```

```python
df.loc[:2] #label-based selection
```

```python
df.iloc[:2] #integer-position based selection
```
