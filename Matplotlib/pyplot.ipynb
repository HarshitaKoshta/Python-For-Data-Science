matplotlib.pyplot is a collection of commoand style functions that make matplotlib work like MATLAB


Each pyplot function makes some chenge to figure e.g. create a figure, creates a plotting area in figure, plots some lines in a plotting area, decorates the plot with labels, etc.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,5,5])  # y axis value
plt.ylabel("Numbers")
plt.xlabel('Indices')
plt.title('MyPlot')
plt.show()
```

```python
plt.plot([1,3,6,9],[2,4,5,13])  # x,y axis value
plt.ylabel("Numbers")
plt.xlabel('Indices')
plt.grid()
plt.show()
```

```python
plt.plot([1,3,6,9],[2,4,5,13],'ro')  # red circle
plt.ylabel("Numbers")
plt.xlabel('Indices')
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(0.,5.,0.2)

#blue dashed, red square and green triangle
plt.plot(t,t**2,'b+-',label='^2')
plt.plot(t,t,'r-*',label='^2.2')
plt.plot(t,t**0.5,'go:',label='^2.5')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
x = [1,2,3,4]
y = [1,8,9,4]
plt.plot(x,y,'go-',linewidth=2)
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [3,4,2,4]
x2 = [9,4,3,2]
y2 = [3,9,3,5]
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='r',linewidth=2.0)
plt.setp(lines[1],'color','g',linewidth = 2.0)
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1= np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)
plt.figure(1)

plt.subplot(211) #2row 1col 1st img
plt.grid()
plt.plot(t1,f(t1),'b-')
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1)
plt.subplot(211)
plt.title('Easy as 1, 2 ,3')
plt.show()
```
