matplotlib.pyplot is a collection of commoand style functions that make matplotlib work like MATLAB


Each pyplot function makes some chenge to figure e.g. create a figure, creates a plotting area in figure, plots some lines in a plotting area, decorates the plot with labels, etc.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel("Numbers")
plt.xlabel('Indices')
plt.title('MyPlot')
plt.show()
```
