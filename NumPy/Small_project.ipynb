## 1️⃣ Dice Roll Simulator (NumPy Random)

Goal: Simulate rolling 2 dice 10,000 times and find probabilities.

```python
import numpy as np


rolls = np.random.randint(1, 7, size=(10000, 2))
sums = rolls.sum(axis=1)

unique, counts = np.unique(sums, return_counts=True)
probabilities = counts / 10000

print(dict(zip(unique, probabilities)))
```

## 3️⃣ Weather Data Simulation

```python
temps = np.random.normal(loc=30, scale=3, size=30)

print("Average:", temps.mean())
print("Max:", temps.max())
print("Min:", temps.min())
```
