### Generator

```python
def my_generator():
    """
    simple genetaor function
    """
    a =1
    print("First Time")
    yield a

    a+=1
    print("Second time")
    yield a

    a+=1
    print("Third time")
    yield a

n = my_generator()
next(n)
```
